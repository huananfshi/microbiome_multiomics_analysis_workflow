```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import  multilabel_confusion_matrix, accuracy_score, classification_report, roc_auc_score,plot_roc_curve
from sklearn.tree import plot_tree
```

```python
#read metabolomics, 
###*_renorm did not work well
plasma_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/plasma_metabolites_log10.csv",index_col=0)
cecal_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/cecal_metabolites_log10.csv",index_col=0)
plasma_metabolites_stats = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_stat_with_pwy_info.csv",index_col=0)
cecal_metabolites_stats = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_stat_with_pwy_info.csv",index_col=0)

plasma_pwy = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/plasma_metabolites_pwy_info.csv",index_col=0)
cecal_pwy = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/cecal_metabolites_pwy_info.csv",index_col=0)
```

```python
plasma_metabolites_con = plasma_metabolites.transpose()[plasma_metabolites.loc["Treatment"]=="Con"]

cecal_metabolites_con = cecal_metabolites.transpose()[cecal_metabolites.loc["Treatment"]=="Con"]
```

```python
#formating plasma data for sklearn
plasma_reads = np.array(plasma_metabolites_con.transpose()[3:].astype(float).transpose())
plasma_metabolites_names = np.array(plasma_metabolites_con.transpose().index[3:])

#encoding groups
plasma_labels_con = np.array(plasma_metabolites_con["Group"])
plasma_lb = preprocessing.LabelEncoder()
plasma_label_con_num = plasma_lb.fit_transform(plasma_labels_con)

n_plasma_samples, n_plasma_features = plasma_reads.shape
n_plasma_classes = plasma_label_con_num.shape[0]
print("plasma metabolomics control: ",n_plasma_samples, "samples; ",n_plasma_features,"features; ",n_plasma_classes,"classes:", plasma_lb.classes_)
```

```python
#formating cecal data for sklearn
cecal_reads = np.array(cecal_metabolites_con.transpose()[3:].astype(float).transpose())
cecal_metabolites_names = np.array(cecal_metabolites_con.transpose().index[3:])

#encoding groups
cecal_labels_con = np.array(cecal_metabolites_con["Group"])
cecal_lb = preprocessing.LabelEncoder()
cecal_label_con_num = cecal_lb.fit_transform(cecal_labels_con)

n_cecal_samples, n_cecal_features = cecal_reads.shape
n_cecal_classes = cecal_label_con_num.shape[0]
print("cecal metabolomics control: ",n_cecal_samples, "samples; ",n_cecal_features,"features; ",n_cecal_classes,"classes:", cecal_lb.classes_)
```

```python
#data sampling: split data for training and testing. 70% for training
plasma_train, plasma_test, plasma_lb_train, plasma_lb_test = train_test_split(plasma_reads,plasma_label_con_num,
                                                                        shuffle = False, test_size = 0.3, random_state = 41)

cecal_train, cecal_test, cecal_lb_train, cecal_lb_test = train_test_split(cecal_reads,cecal_label_con_num,
                                                                        shuffle = True, test_size = 0.3, random_state = 41)
```

```python
#building random forest classifier, train/fit with train data

plasma_forest = ExtraTreesClassifier(random_state=1,n_estimators=3000)
plasma_classfier = plasma_forest.fit(plasma_train, plasma_lb_train)

cecal_forest = RandomForestClassifier(random_state=1,n_estimators=3000)
cecal_classfier = cecal_forest.fit(cecal_train, cecal_lb_train)
```

```python
#Model Evaluation

#confusion matrix and prediction accuracy
#calculate predicted data with test input/reads
plasma_pred = plasma_forest.predict(plasma_test)
cecal_pred = cecal_forest.predict(cecal_test)
#calculate confusion matrix 
cm_plasma = multilabel_confusion_matrix(plasma_lb_test, plasma_pred)
cm_cecal = multilabel_confusion_matrix(cecal_lb_test, cecal_pred)
print("plasma metabolomics \n", cm_plasma, "\n cecal metabolomics \n", cm_cecal)
#accuracy with test data
plasma_accuracies = accuracy_score(plasma_lb_test, plasma_pred)
plasma_auc = roc_auc_score(plasma_lb_test, plasma_pred)
cecal_accuracies = accuracy_score(cecal_lb_test, cecal_pred)
cecal_auc = roc_auc_score(cecal_lb_test, cecal_pred)
print("Model test accuracy for plasma: {0:0.2f}%; AUC: {1:0.2f}".format(plasma_accuracies*100,plasma_auc))
print("Model test accuracy for cecal: {0:0.2f}%;AUC: {1:0.2f}".format(cecal_accuracies*100,cecal_auc))
```

```python
#classification reports
print("plasma metabolomics classification report \n",classification_report(plasma_lb_test, plasma_pred, target_names=plasma_lb.classes_,zero_division=0))
print("cecal metabolomics classification report \n",classification_report(cecal_lb_test, cecal_pred, target_names=cecal_lb.classes_,zero_division=0))
```

```python
# 5 fold CV accuracy with training data
plasma_cv_accuracy = np.mean(cross_val_score(plasma_classfier, plasma_train,plasma_lb_train, cv =2))
cecal_cv_accuracy = np.mean(cross_val_score(cecal_classfier, cecal_train,cecal_lb_train, cv =2))
print("Model 5-fold CV accuracy for plasma: {0:0.2f}%.".format(plasma_cv_accuracy*100))
print("Model 5-fold CV accuracy for cecal: {0:0.2f}%.".format(cecal_cv_accuracy*100))
```

```python
#ROC curves and AUC with testing data
#roc_curve only works with binary array, transform y_test to binary

#plot and save ROC curve    
plasma_roc = plt.figure(0)
plot_roc_curve(plasma_classfier,plasma_test,plasma_lb_test)
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Plasma test ROC")


cecal_roc = plt.figure(2)
plot_roc_curve(cecal_classfier,cecal_test,cecal_lb_test)

plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Cecal test ROC")

plasma_roc.savefig("/Users/huananshi/Box Sync/EODF_analysis/outputs/Plasma_RF_test_con_roc.png",format="png")
cecal_roc.savefig("/Users/huananshi/Box Sync/EODF_analysis/outputs/Cecal_RF_test_con_roc.png",format="png")
```

```python
#get feature importances from RF
plasma_importance = pd.DataFrame(index=plasma_metabolites_names,columns=["importance"])
cecal_importance = pd.DataFrame(index=cecal_metabolites_names,columns=["importance"])
plasma_importance['importance'] = plasma_forest.feature_importances_
cecal_importance['importance'] = cecal_forest.feature_importances_
#save RF feature importances
plasma_importance.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importances_accuracy_{0:0.0f}%.csv".format(plasma_accuracies*100))
cecal_importance.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importances_accuracy_{0:0.0f}%.csv".format(cecal_accuracies*100))

#RF feature importances with metabolites pathway information
plasma_importance_pwy= pd.concat([plasma_pwy,plasma_importance],axis = 1)
cecal_importance_pwy = pd.concat([cecal_pwy,cecal_importance],axis = 1)
plasma_importance_pwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importances_with_pwy_accuracy_{0:0.0f}%.csv".format(plasma_accuracies*100))
cecal_importance_pwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importances_with_pwy_accuracy_{0:0.0f}%.csv".format(cecal_accuracies*100))

#RF feature importances groupby sub pathways
plasma_importance_subpwy = plasma_importance_pwy.groupby("SUB PATHWAY").sum()
cecal_importance_subpwy = cecal_importance_pwy.groupby("SUB PATHWAY").sum()
plasma_importance_subpwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importances_by_subpwy_accuracy_{0:0.0f}%.csv".format(plasma_accuracies*100))
cecal_importance_subpwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importances_by_subpwy_accuracy_{0:0.0f}%.csv".format(cecal_accuracies*100))

```

```python
#calculate feature importance by classes
def class_feature_importance(X, Y, feature_importances):
    N, M = X.shape
    X = preprocessing.scale(X)

    out = {}
    for c in set(Y):
        out[c] = dict(
            zip(range(M), np.mean(X[Y==c, :], axis=0)*feature_importances)
        )

    return out
```

```python
#calculate feature importance by classes
plasma_importance_by_class = class_feature_importance(plasma_reads, plasma_labels_con, plasma_classfier.feature_importances_)
cecal_importance_by_class = class_feature_importance(cecal_reads, cecal_labels_con, cecal_classfier.feature_importances_)
plasma_importance_by_class = pd.DataFrame(plasma_importance_by_class)
plasma_importance_by_class.index = plasma_metabolites_names
cecal_importance_by_class = pd.DataFrame(cecal_importance_by_class)
cecal_importance_by_class.index = cecal_metabolites_names
plasma_importance_by_class.to_csv(
    "/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importances_by_class_accuracy_{0:0.0f}%.csv"
    .format(plasma_accuracies*100))

cecal_importance_by_class.to_csv(
    "/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importances_by_class_accuracy_{0:0.0f}%.csv"
    .format(cecal_accuracies*100))

#RF feature importance_by_class with metabolites pathway information
plasma_importance_by_class_pwy= pd.concat([plasma_pwy,plasma_importance_by_class],axis = 1)
cecal_importance_by_class_pwy = pd.concat([cecal_pwy,cecal_importance_by_class],axis = 1)
plasma_importance_by_class_pwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importance_by_classs_with_pwy_accuracy_{0:0.0f}%.csv".format(plasma_accuracies*100))
cecal_importance_by_class_pwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importance_by_classs_with_pwy_accuracy_{0:0.0f}%.csv".format(cecal_accuracies*100))

#RF feature importance_by_class groupby sub pathways
plasma_importance_by_class_subpwy = plasma_importance_by_class_pwy.groupby("SUB PATHWAY").sum()
cecal_importance_by_class_subpwy = cecal_importance_by_class_pwy.groupby("SUB PATHWAY").sum()
plasma_importance_by_class_subpwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_con_RF_feature_importance_by_classs_by_subpwy_accuracy_{0:0.0f}%.csv".format(plasma_accuracies*100))
cecal_importance_by_class_subpwy.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_con_RF_feature_importance_by_classs_by_subpwy_accuracy_{0:0.0f}%.csv".format(cecal_accuracies*100))

```
