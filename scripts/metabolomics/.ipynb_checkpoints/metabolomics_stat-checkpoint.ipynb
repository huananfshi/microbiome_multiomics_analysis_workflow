```python
import pandas as pd
import scipy as sp
import numpy as np
from statsmodels.formula.api import ols
import statsmodels.stats.multicomp
from statsmodels.stats.anova import anova_lm
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def metabolomics_stat(result):
    #there are a ton of stuff can be simplified or optimized for more general use: I can load the file and filter groups based on given names
    #i can set an argument to determine the numbers of groups, factors and determine which stats should be run
    res_sca=result  
    res_sca_t = res_sca.transpose()
    metabolites = res_sca_t.columns.tolist()[3:]
    res_stat = pd.DataFrame(index=metabolites)
    res_stat['p_interaction']=""
    res_stat['p_genotype']=""
    res_stat['p_treatment']=""
    res_stat['p_WC_SC']=""
    res_stat['p_WC_SE']=""
    res_stat['p_WC_WE']=""
    res_stat['p_WE_SC']=""
    res_stat['p_WE_SE']=""
    res_stat['p_SC_SE']=""
    res_stat['log2fc_SC_WC']=""
    res_stat['log2fc_SC_SE']=""
    for metabolite in metabolites: 
        test = res_sca_t.loc[:,["Group","Genotype","Treatment",metabolite]]
        test.columns = ["Group","Genotype","Treatment","value"]
        test.loc[:,"value"]=test.loc[:,"value"].astype(float)
        model = ols('value ~ C(Genotype)*C(Treatment)', test).fit()
        aov_table = anova_lm(model, typ=2)
        p_genotype = aov_table["PR(>F)"][0]
        p_treatment = aov_table["PR(>F)"][1]
        p_interaction = aov_table["PR(>F)"][2]
        res_stat.loc[metabolite,'p_genotype']=p_genotype
        res_stat.loc[metabolite,'p_treatment']=p_treatment
        res_stat.loc[metabolite,'p_interaction']= p_interaction
        if p_genotype>1 and p_treatment>1 and p_interaction>1:
            res_stat.loc[metabolite,['p_WC_SC','p_WC_SE','p_WC_WE','p_WE_SC','p_WE_SE',"p_SC_SE"]]\
            =[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
            print(metabolite+" two way ANOVA ns")
        else:
            mc = statsmodels.stats.multicomp.MultiComparison(test['value'],test['Group'])
            mc_HS_table = mc.allpairtest(stats.ttest_ind,method='fdr_bh')[0]
            res_stat.loc[metabolite,['p_WC_SC','p_WC_SE','p_WC_WE','p_WE_SC','p_WE_SE',"p_SC_SE"]]\
            = [mc_HS_table.data[2][4]]+[mc_HS_table.data[4][4]]+[mc_HS_table.data[6][4]]\
            +[mc_HS_table.data[3][4]]+[mc_HS_table.data[5][4]]+[mc_HS_table.data[1][4]]
            print(metabolite+" Multicomp (FDR_BH) EODF")
        print(metabolite + "  done EODF")
        
    #indices = res_stat.index[(res_stat.p_WC_SC<=0.05) & (res_stat.p_WC_SE>0.05) &( res_stat.p_SC_SE<=0.05)]
    #indices = ["Group","Genotype","Treatment"]+indices.tolist()
    #res_sca_sig= res_sca.loc[indices,:]
    return res_stat #,res_sca_sig
```

```python
plasma_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/plasma_metabolites_log10.csv",index_col=0)
cecal_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/cecal_metabolites_log10.csv",index_col=0)
plasma_pwy = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/plasma_metabolites_pwy_info.csv",index_col=0)
cecal_pwy = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/cecal_metabolites_pwy_info.csv",index_col=0)
```

```python
plasma_metabolites_stat = metabolomics_stat(plasma_metabolites)
```

```python
#calculating fold change (log2)
plasma_metabolites_t = plasma_metabolites.transpose()
metabolites = plasma_metabolites_t.columns.tolist()[3:]
plasma_metabolites_t[metabolites]=plasma_metabolites_t[metabolites].astype(float)
plasma_metabolites_t[metabolites] = plasma_metabolites_t[metabolites].applymap(lambda x: 10**x)
plasma_metabolites_sum = plasma_metabolites_t.groupby("Group").mean().transpose()
plasma_metabolites_stat['log2fc_SC_WC'] = plasma_metabolites_sum["SHRSP Con"].div(plasma_metabolites_sum["WKY Con"]).apply(np.log2)
plasma_metabolites_stat['log2fc_SC_SE'] = plasma_metabolites_sum["SHRSP Con"].div(plasma_metabolites_sum["SHRSP EODF"]).apply(np.log2)
```

```python
cecal_metabolites_stat = metabolomics_stat(cecal_metabolites)
```

```python
#calculating fold change (log2)
cecal_metabolites_t = cecal_metabolites.transpose()
metabolites = cecal_metabolites_t.columns.tolist()[3:]
cecal_metabolites_t[metabolites]=cecal_metabolites_t[metabolites].astype(float)
cecal_metabolites_t[metabolites] = cecal_metabolites_t[metabolites].applymap(lambda x: 10**x)
cecal_metabolites_sum = cecal_metabolites_t.groupby("Group").mean().transpose()
cecal_metabolites_stat['log2fc_SC_WC'] = cecal_metabolites_sum["SHRSP Con"].div(cecal_metabolites_sum["WKY Con"]).apply(np.log2)
cecal_metabolites_stat['log2fc_SC_SE'] = cecal_metabolites_sum["SHRSP Con"].div(cecal_metabolites_sum["SHRSP EODF"]).apply(np.log2)
```

```python
plasma_metabolites_stat.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/plasma_metabolites_stat.csv")
cecal_metabolites_stat.to_csv("/Users/huananshi/Box Sync/EODF_analysis/outputs/cecal_metabolites_stat.csv")
```
