```python
import numpy as np
import matplotlib as plt
import pandas as pd
from sklearn import preprocessing
import seaborn as sns
```

```python
plasma_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/plasma_metabolites_log10.csv", index_col=0)
cecal_metabolites = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/metabolomics/cecal_metabolites_log10.csv", index_col=0)
gene_abun = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/genes/ecs_relab_all_nonspecies_name.csv", index_col=0)
pathway_abun = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/pathways/pathways_relab_all_nonspecies.csv", index_col=0)
species = pd.read_csv("/Users/huananshi/Box Sync/EODF_analysis/intermediate_files/taxonomy/taxonomic_profiles_all_species.csv",index_col=0)
```

```python
gene_abun_num = np.array(gene_abun[1:])
pathway_abun_num = np.array(pathway_abun)
plasma_metabolites_t_num = np.array(plasma_metabolites[4:].transpose())
cecal_metabolites_t_num = np.array(cecal_metabolites[4:].transpose())
```

```python
gene_abun_renorm = pd.DataFrame(preprocessing.StandardScaler().fit_transform(gene_abun[1:]), 
                                index= gene_abun.index[1:], columns= gene_abun.columns)
pathway_abun_renorm = pd.DataFrame(preprocessing.StandardScaler().fit_transform(pathway_abun), 
                                   index= pathway_abun.index, columns= pathway_abun.columns)
gene_abun_renorm.loc["Gene Name"] = gene_abun.loc["Gene Name"]
gene_abun_renorm = gene_abun_renorm.reindex(gene_abun.index)
species_renorm = pd.DataFrame(preprocessing.PowerTransformer(method="yeo-johnson").fit_transform\
                              (preprocessing.StandardScaler().fit_transform(species)), 
                              index= species.index, columns= species.columns)
```
